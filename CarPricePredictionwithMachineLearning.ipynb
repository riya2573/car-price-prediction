```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('car data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import datetime
date_time = datetime.datetime.now()
df['Age']=date_time.year - df['Year']
df.drop('Year', axis=1, inplace=True)
df.head()
```

```python
sns.boxplot(df['Selling_Price'])
```

```python
sorted(df['Selling_Price'], reverse=True)
```

```python
df = df[~(df['Selling_Price']>=33.0) & (df['Selling_Price']<=35.0)]
```

```python
df.shape
```

```python
df['Fuel_Type'].unique()
```

```python
df['Fuel_Type'] = df['Fuel_Type'].map({'Petrol':0, 'Diesel':1, 'CNG':2})
df['Fuel_Type'].unique()
```

```python
df['Selling_type'].unique()
df['Selling_type'] = df['Selling_type'].map({'Dealer':0, 'Individual':1})
```

```python
df['Selling_type'].unique()
```

```python
df['Transmission'].unique()
df['Transmission'] = df['Transmission'].map({'Manual':0, 'Automatic':1})
```

```python
df['Transmission'].unique()
```

```python
df.head()
```

```python
X = df.drop(['Car_Name', 'Selling_Price'], axis=1)
Y = df['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
xgb = GradientBoostingRegressor()
xgb.fit(x_train, y_train)
```

```python
xg = XGBRegressor()
xg.fit(x_train, y_train)
```

```python
y_pred1 = lr.predict(x_test)
y_pred2 = rf.predict(x_test)
y_pred3 = xgb.predict(x_test)
y_pred4 = xg.predict(x_test)
```

```python
from sklearn import metrics

score1 = metrics.r2_score(y_test, y_pred1)
score2 = metrics.r2_score(y_test, y_pred2)
score3 = metrics.r2_score(y_test, y_pred3)
score4 = metrics.r2_score(y_test, y_pred4)
```

```python
print(score1, score2, score3, score4)
```

```python
final_df = pd.DataFrame({'Models':['LR', 'RF', 'GBR', 'XG'],
                        "R2_SCORE":[score1, score2, score3, score4]})
final_df
```

```python
sns.barplot(x='Models', y='R2_SCORE', data=final_df)
```

```python
xg = XGBRegressor()
xg_final = xg.fit(X, Y)
```

```python
import joblib

joblib.dump(xg_final, 'car_price_predictor')
```

```python
model = joblib.load('car_price_predictor')
```
